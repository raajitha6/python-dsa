```python
#scheduling
import random
a=int(input('Enter no. of teams: '))
teams=[]
for i in range(a):
    t=input('Team name: ')
    teams.append(t)
m=int(input('No. of meet b/w teams: '))

matches=[]
for i in range(0,a-1):
    for j in range(i+1,a):
        for k in range(m):
            matches.append([teams[i],teams[j]])
            
random.shuffle(matches)
pos=1
for i in matches:
    print('MATCH {} : {} vs {}'.format(pos,i[0],i[1]))
    pos+=1
```

#### LINKED LIST :-

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class LinkedList:
    def __init__(self):
        self.head=None

    def insert(self,data):
        newnode=Node(data)
        if self.head==None:
            self.head=newnode
        else:
            temp=self.head
            while temp.next:
                temp=temp.next
            temp.next=newnode

    def display(self):
        temp=self.head
        while temp:
            print(temp.data,"->",end=" ")
            temp=temp.next
    

l=LinkedList()
l.insert(100)
l.insert(200)
l.insert(300)
l.insert(400)
l.insert(500)
l.display()
```

```python
class Stack:
    def __init__(self):
        self.l=[]
    def push(self,data):
        self.l.append(data)
    def display(self):
        b='->'.join(map(str,self.l))
        print(b)
    def pop(self):
        self.l.pop()
    def peek(self):
        print(self.l[-1])



s=Stack()
s.push(100)
s.push(200)
s.push(89)
s.display()
s.pop()
s.peek()
```

```python
class Graph:
    def __init__(self,v):
        self.adj=[[] for i in range(v)]

    def add_edge(self,a,b):
        self.adj[a].append(b)
        self.adj[b].append(a)
        
    def display(self):
        for i in range(len(self.adj)):
            print('Node:',i,'->',end=" ")
            #for j in self.adj[i]:
            #    print(j,",",end=" ")
            #print()
            res=",".join(map(str,self.adj[i]))
            print(res)

g=Graph(6)
g.add_edge(0,2)
g.add_edge(0,3)
g.add_edge(1,4)
g.add_edge(1,3)
g.add_edge(3,2)
g.add_edge(3,3)
g.add_edge(5,1)
g.display()
```
